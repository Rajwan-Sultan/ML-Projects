```python
# Creating first MLP in keras
from pandas import read_csv
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split


#? This will not work on the local machine because the tensorflow is not installed on the local machine

#1 Loading data

filename = '/content/gdrive/MyDrive/ML/pima-indians-diabetes.csv' # this is my file directory where the pima indian diabetes dataset is located
names = ['preg','plas','pres','skin','test','mass','pedi','age','class']
df = read_csv(filename,names = names)
array = df.values # Making the data frame into an array of values
X = array[:,0:8] # this will take all the rows of the array and take the columns upto 7
Y = array[:,8] # this will take all the rows and take only the 8th column
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,random_state=42)


#2 Create / define model
model = Sequential()
model.add(Dense(12,input_dim=8,activation = 'relu'))# 1st Hidden Layer  # input dimension is given according to the given features
model.add(Dense(8,activation='relu'))# 2nd Hidden Layer
model.add(Dense(1,activation='sigmoid'))# Output Layer

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy']) # loss function , optimizer(gradient decent algorithm) , what kind of evaluation


model.fit(X_train,Y_train,epochs=150,batch_size=10) # fitting the values in the model

score = model.evaluate(X_test,Y_test) # evaluating the score of the model

print("%s: %.2f%%"%(model.metrics_names[0],score[0]*100)) # loss

print("%s: %.2f%%"%(model.metrics_names[1],score[1]*100)) # accuracy

```
